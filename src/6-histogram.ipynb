```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt
import numpy as np
from typing import Union, List, Optional, Tuple
import warnings

def validate_histogram_data(data: Union[List, np.ndarray], data_name: str = "data") -> np.ndarray:
    """
    Comprehensive validation for histogram data.
    
    Parameters:
    -----------
    data : list, tuple, or numpy array
        Input data for histogram
    data_name : str
        Name of data for error messages
        
    Returns:
    --------
    np.ndarray : Cleaned data array
    """
    if data is None:
        raise ValueError(f"{data_name} cannot be None")
    
    if not isinstance(data, (list, tuple, np.ndarray)):
        raise TypeError(f"{data_name} must be list, tuple, or numpy array, got {type(data)}")
    
    if len(data) == 0:
        raise ValueError(f"{data_name} cannot be empty")
    
    try:
        # Convert to numpy array
        data_array = np.array(data, dtype=float)
        
        # Handle NaN values
        nan_count = np.sum(np.isnan(data_array))
        if nan_count > 0:
            print(f"Warning: {nan_count} NaN values detected and will be removed")
            data_array = data_array[~np.isnan(data_array)]
            if len(data_array) == 0:
                raise ValueError("No valid data remains after removing NaN values")
        
        # Handle infinite values
        inf_count = np.sum(np.isinf(data_array))
        if inf_count > 0:
            print(f"Warning: {inf_count} infinite values detected and will be removed")
            data_array = data_array[~np.isinf(data_array)]
            if len(data_array) == 0:
                raise ValueError("No valid data remains after removing infinite values")
        
        return data_array
        
    except (ValueError, TypeError) as e:
        if "could not convert" in str(e):
            raise ValueError(f"Cannot convert {data_name} to numeric array. Check for non-numeric values.")
        raise

def validate_bins_parameter(bins: Union[int, str, np.ndarray, List], data_length: int) -> Union[int, str, np.ndarray]:
    """
    Validate bins parameter for histogram.
    
    Parameters:
    -----------
    bins : int, str, array-like, or sequence
        Bins specification
    data_length : int
        Length of data array
        
    Returns:
    --------
    Valid bins parameter
    """
    if bins is None:
        return 'auto'  # Default to auto binning
    
    if isinstance(bins, str):
        valid_methods = ['auto', 'fd', 'scott', 'stone', 'rice', 'sturges', 'sqrt']
        if bins.lower() not in valid_methods:
            raise ValueError(f"Invalid binning method '{bins}'. Valid methods: {valid_methods}")
        return bins.lower()
    
    if isinstance(bins, int):
        if bins <= 0:
            raise ValueError("Number of bins must be positive")
        if bins > data_length:
            print(f"Warning: Number of bins ({bins}) exceeds data points ({data_length}). Using {data_length} bins.")
            return data_length
        return bins
    
    if isinstance(bins, (list, tuple, np.ndarray)):
        bins_array = np.array(bins)
        if len(bins_array) < 2:
            raise ValueError("Bin edges array must have at least 2 elements")
        if not np.all(np.diff(bins_array) > 0):
            raise ValueError("Bin edges must be monotonically increasing")
        return bins_array
    
    raise TypeError(f"Invalid bins type: {type(bins)}. Must be int, str, or array-like")

def calculate_optimal_bins(data: np.ndarray, method: str = 'auto') -> int:
    """
    Calculate optimal number of bins using various methods.
    
    Parameters:
    -----------
    data : np.ndarray
        Input data
    method : str
        Binning method
        
    Returns:
    --------
    int : Optimal number of bins
    """
    n = len(data)
    
    if method == 'sturges':
        return int(np.ceil(np.log2(n) + 1))
    elif method == 'rice':
        return int(np.ceil(2 * (n ** (1/3))))
    elif method == 'sqrt':
        return int(np.ceil(np.sqrt(n)))
    elif method == 'fd':  # Freedman-Diaconis
        iqr = np.percentile(data, 75) - np.percentile(data, 25)
        if iqr == 0:
            return int(np.ceil(np.sqrt(n)))  # Fallback
        bin_width = 2 * iqr * (n ** (-1/3))
        return int(np.ceil((np.max(data) - np.min(data)) / bin_width))
    else:  # auto, scott, stone
        # Use matplotlib's internal binning
        return 'auto'

def create_enhanced_histogram(data=None, bins=5, figsize=(8, 5), color='#6B5B95', 
                            title='Enhanced Histogram of Data', xlabel='Value', 
                            ylabel='Frequency', density=False, alpha=0.8,
                            show_stats=True, show_kde=False):
    """
    Create an enhanced histogram with comprehensive error handling and validation.
    
    Parameters:
    -----------
    data : array-like
        Input data for histogram
    bins : int, str, or array-like
        Bins specification
    figsize : tuple
        Figure size
    color : str
        Histogram color
    title : str
        Plot title
    xlabel : str
        X-axis label
    ylabel : str
        Y-axis label
    density : bool
        If True, draw density instead of frequency
    alpha : float
        Transparency level
    show_stats : bool
        Whether to display statistics on plot
    show_kde : bool
        Whether to overlay KDE curve
        
    Returns:
    --------
    dict : Results including success status and statistics
    """
    
    results = {
        'success': False,
        'statistics': {},
        'bin_info': {},
        'warnings': []
    }
    
    try:
        # Default data if none provided
        if data is None:
            data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
            print("Using default data")
        
        # Validate figsize
        if not isinstance(figsize, (tuple, list)) or len(figsize) != 2:
            raise ValueError("figsize must be a tuple or list of length 2")
        if figsize[0] <= 0 or figsize[1] <= 0:
            raise ValueError("figsize dimensions must be positive")
        
        # Validate alpha
        if not 0 <= alpha <= 1:
            raise ValueError("alpha must be between 0 and 1")
        
        print("Validating histogram data...")
        
        # Data validation
        data_clean = validate_histogram_data(data, "input data")
        original_length = len(data) if hasattr(data, '__len__') else 0
        cleaned_length = len(data_clean)
        
        if cleaned_length != original_length:
            results['warnings'].append(f"Removed {original_length - cleaned_length} invalid data points")
        
        # Validate bins parameter
        bins_validated = validate_bins_parameter(bins, len(data_clean))
        
        # Calculate statistics
        stats = {
            'count': len(data_clean),
            'mean': np.mean(data_clean),
            'std': np.std(data_clean),
            'min': np.min(data_clean),
            'max': np.max(data_clean),
            'median': np.median(data_clean),
            'q25': np.percentile(data_clean, 25),
            'q75': np.percentile(data_clean, 75)
        }
        
        # Check for data distribution warnings
        if stats['std'] == 0:
            results['warnings'].append("All data points are identical - histogram may not be informative")
        
        if stats['count'] < 5:
            results['warnings'].append(f"Small sample size ({stats['count']}) - consider more data for reliable histogram")
        
        # Calculate optimal bins if using string method
        if isinstance(bins_validated, str):
            optimal_bins = calculate_optimal_bins(data_clean, bins_validated)
            if isinstance(optimal_bins, int):
                bins_validated = optimal_bins
                results['bin_info']['method'] = bins
                results['bin_info']['calculated_bins'] = optimal_bins
        
        print(f"Data validated: {stats['count']} points, range [{stats['min']:.2f}, {stats['max']:.2f}]")
        
        # Create the plot
        plt.figure(figsize=figsize)
        
        # Create histogram
        n, bin_edges, patches = plt.hist(data_clean, bins=bins_validated, color=color, 
                                        edgecolor='black', alpha=alpha, density=density,
                                        linewidth=0.5)
        
        # Store bin information
        results['bin_info']['n_bins'] = len(bin_edges) - 1
        results['bin_info']['bin_width'] = np.mean(np.diff(bin_edges))
        results['bin_info']['bin_edges'] = bin_edges.tolist()
        
        # Add KDE overlay if requested and data is suitable
        if show_kde and len(data_clean) > 3 and stats['std'] > 0:
            try:
                from scipy.stats import gaussian_kde
                kde = gaussian_kde(data_clean)
                x_range = np.linspace(stats['min'], stats['max'], 100)
                kde_values = kde(x_range)
                
                if density:
                    plt.plot(x_range, kde_values, 'r-', linewidth=2, label='KDE', alpha=0.8)
                else:
                    # Scale KDE to match frequency histogram
                    kde_scaled = kde_values * len(data_clean) * results['bin_info']['bin_width']
                    plt.plot(x_range, kde_scaled, 'r-', linewidth=2, label='KDE', alpha=0.8)
                
                plt.legend()
            except ImportError:
                results['warnings'].append("scipy not available - KDE overlay skipped")
            except Exception as e:
                results['warnings'].append(f"KDE calculation failed: {e}")
        
        # Enhance plot appearance
        plt.title(title, fontsize=12, fontweight='bold', pad=15)
        plt.xlabel(xlabel, fontsize=10)
        
        # Adjust ylabel based on density setting
        if density:
            plt.ylabel('Density', fontsize=10)
        else:
            plt.ylabel(ylabel, fontsize=10)
        
        plt.grid(axis='y', linestyle='--', alpha=0.5, color='gray')
        
        # Add statistics text box if requested
        if show_stats:
            stats_text = f'Count: {stats["count"]}\n'
            stats_text += f'Mean: {stats["mean"]:.2f}\n'
            stats_text += f'Std: {stats["std"]:.2f}\n'
            stats_text += f'Range: [{stats["min"]:.2f}, {stats["max"]:.2f}]'
            
            plt.text(0.98, 0.98, stats_text, transform=plt.gca().transAxes,
                    verticalalignment='top', horizontalalignment='right',
                    bbox=dict(boxstyle='round', facecolor='white', alpha=0.8),
                    fontsize=9, fontfamily='monospace')
        
        # Add value labels on bars for small number of bins
        if results['bin_info']['n_bins'] <= 10 and not density:
            for i, (patch, count) in enumerate(zip(patches, n)):
                if count > 0:  # Only label non-empty bins
                    plt.text(patch.get_x() + patch.get_width()/2, patch.get_height() + 0.1,
                            f'{int(count)}', ha='center', va='bottom', fontsize=9)
        
        plt.tight_layout()
        plt.show()
        
        # Store results
        results['success'] = True
        results['statistics'] = stats
        
        # Print summary
        print(f"\n=== HISTOGRAM SUMMARY ===")
        print(f"Data points: {stats['count']}")
        print(f"Bins used: {results['bin_info']['n_bins']}")
        print(f"Mean: {stats['mean']:.3f} ± {stats['std']:.3f}")
        print(f"Range: [{stats['min']:.3f}, {stats['max']:.3f}]")
        print(f"Median: {stats['median']:.3f}")
        
        if results['warnings']:
            print(f"\nWarnings:")
            for warning in results['warnings']:
                print(f"- {warning}")
        
        print("Histogram created successfully!")
        
        return results
        
    except ValueError as ve:
        print(f"Validation Error: {ve}")
        results['error'] = str(ve)
        return results
    except TypeError as te:
        print(f"Type Error: {te}")
        results['error'] = str(te)
        return results
    except Exception as e:
        print(f"Unexpected error: {e}")
        results['error'] = str(e)
        return results
    finally:
        # Clean up
        plt.close('all')

# Example usage and testing
if __name__ == "__main__":
    print("Creating enhanced histogram...")
    
    # Test with original data
    original_data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
    result1 = create_enhanced_histogram(original_data, bins=5)
    
    print("\n" + "="*60)
    print("COMPREHENSIVE ERROR HANDLING TESTS")
    print("="*60)
    
    # Test cases for various error conditions
    test_cases = [
        {
            'name': 'Empty data array',
            'data': [],
            'expected': 'ValueError'
        },
        {
            'name': 'None data',
            'data': None,
            'expected': 'ValueError'
        },
        {
            'name': 'Data with NaN values',
            'data': [1, 2, np.nan, 4, 5, np.nan],
            'expected': 'Warning'
        },
        {
            'name': 'Data with infinite values',
            'data': [1, 2, np.inf, 4, 5],
            'expected': 'Warning'
        },
        {
            'name': 'Invalid bins parameter (negative)',
            'data': [1, 2, 3, 4, 5],
            'bins': -5,
            'expected': 'ValueError'
        },
        {
            'name': 'Invalid bins parameter (zero)',
            'data': [1, 2, 3, 4, 5],
            'bins': 0,
            'expected': 'ValueError'
        },
        {
            'name': 'String data (non-numeric)',
            'data': ['a', 'b', 'c'],
            'expected': 'ValueError'
        },
        {
            'name': 'Mixed valid/invalid data',
            'data': [1, 2, 'invalid', 4, 5],
            'expected': 'ValueError'
        },
        {
            'name': 'Single data point',
            'data': [5],
            'expected': 'Warning'
        },
        {
            'name': 'All identical values',
            'data': [3, 3, 3, 3, 3],
            'expected': 'Warning'
        },
        {
            'name': 'Invalid figsize',
            'data': [1, 2, 3, 4, 5],
            'figsize': (-5, 10),
            'expected': 'ValueError'
        },
        {
            'name': 'Invalid alpha value',
            'data': [1, 2, 3, 4, 5],
            'alpha': 1.5,
            'expected': 'ValueError'
        }
    ]
    
    for i, test_case in enumerate(test_cases, 1):
        print(f"\nTest {i}: {test_case['name']}")
        print("-" * 40)
        
        try:
            # Prepare parameters
            params = {
                'data': test_case['data']
            }
            
            # Add optional parameters if they exist in test case
            for param in ['bins', 'figsize', 'alpha']:
                if param in test_case:
                    params[param] = test_case[param]
            
            # Run test
            result = create_enhanced_histogram(**params)
            
            if result['success']:
                print("✓ Test completed successfully")
                if result['warnings']:
                    print(f"  Warnings: {len(result['warnings'])}")
            else:
                print(f"✗ Test failed as expected: {result.get('error', 'Unknown error')}")
                
        except Exception as e:
            print(f"✗ Caught expected error: {type(e).__name__}: {e}")
    
    print("\n" + "="*60)
    print("ADVANCED FUNCTIONALITY TESTS")
    print("="*60)
    
    # Test with different binning methods
    test_data = np.random.normal(50, 15, 100)  # Normal distribution
    
    binning_methods = ['auto', 'sturges', 'rice', 'sqrt', 'fd']
    
    for method in binning_methods:
        print(f"\nTesting binning method: {method}")
        try:
            result = create_enhanced_histogram(
                data=test_data, 
                bins=method,
                title=f'Histogram with {method} binning',
                show_stats=True,
                show_kde=True if method == 'auto' else False
            )
            if result['success']:
                print(f"✓ {method} method: {result['bin_info']['n_bins']} bins used")
        except Exception as e:
            print(f"✗ {method} method failed: {e}")
    
    # Test with density plot
    print(f"\nTesting density histogram...")
    result = create_enhanced_histogram(
        data=test_data,
        bins=20,
        density=True,
        title='Density Histogram',
        show_kde=True,
        alpha=0.6
    )
    
    print("\nAll tests completed!")
    print("Enhanced histogram code is ready for use!")
    
    # Usage examples
    print("\n" + "="*60)
    print("USAGE EXAMPLES")
    print("="*60)
    print("""
# Basic usage:
result = create_enhanced_histogram([1,2,2,3,3,3,4,4,5])

# Advanced usage:
result = create_enhanced_histogram(
    data=your_data,
    bins='auto',           # or specific number like 20
    figsize=(10, 6),      # custom figure size
    color='steelblue',    # custom color
    title='My Data Distribution',
    density=True,         # show density instead of frequency
    show_stats=True,      # display statistics box
    show_kde=True,        # overlay KDE curve
    alpha=0.7            # transparency
)

# Check results:
if result['success']:
    print(f"Mean: {result['statistics']['mean']:.2f}")
    print(f"Std: {result['statistics']['std']:.2f}")
    print(f"Bins used: {result['bin_info']['n_bins']}")
""")
```
